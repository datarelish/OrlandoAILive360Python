# Data Visualization using Python


In this tutorial, the diamonds dataset will be used in order to demonstrate different ways of visualizing data using Pandas and Seaborn.


# Import the Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import altair as alt
```

# Import the Data

Import the Diamonds dataset from github.

Set the diamonds dataset to the diamonds variable.

```python
diamonds = pd.read_csv("https://raw.githubusercontent.com/datarelish/datasets/master/diamonds.csv")
```

Read out the contents of the diamonds dataset.

```python
diamonds
```

Set the colours for the chart, and assign to the colourscheme variable.

```python
# Specify and set 20 Colors
colourscheme = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),  
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),  
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),  
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),  
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]
             

# Rescale to values between 0 and 1 
for i in range(len(colourscheme)):  
    r, g, b = colourscheme[i]  
    colourscheme[i] = (r / 255., g / 255., b / 255.)
    
```

Plot the diamonds dataset in a scatter chart.

```python
diamonds.plot(kind='scatter', x='carat', y='price', ylim=((0, 25000)), xlim=((0., 5)), s=diamonds['price']*0.0001)
```

Let's see the data grouped by the cut of the diamond.

```python
# Set up the grouping
fig, ax = plt.subplots()
groups = diamonds.groupby('cut')
colors = colourscheme[::2]

# Plotting each group 
for i, (name, group) in enumerate(groups):
    group.plot(kind='scatter', x='carat', y='price', ylim=((0, 25000)), xlim=((0., 5)),
               s=10+group['carat'], 
               label=name, ax=ax, color=colors[i])

# Legend for State colours
lgd = ax.legend(numpoints=1, loc=1, borderpad=1, 
            frameon=True, framealpha=0.9, title="Cut")
for handle in lgd.legendHandles:
    handle.set_sizes([100.0])
```

```python
sizes = [1, 25, 75, 100] 
marker_size = pd.cut(diamonds['price'], range(0, 2500, 500), labels=sizes) 
sns.lmplot('carat', 'price', data=diamonds, hue='cut', fit_reg=False, scatter_kws={'s':marker_size})
plt.ylim((0, 25000))
```
